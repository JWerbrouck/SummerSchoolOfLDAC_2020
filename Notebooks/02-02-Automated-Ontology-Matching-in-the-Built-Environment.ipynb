# Automated ontology matching in the built environment
Lecturer: **Georg Ferdinand Schneider**, [RG](https://www.researchgate.net/profile/Georg_Schneider3)  

### Problem Description

Ontology-based modelling in the Architecture, Engineering, Construction/ Facility Management (AEC/FM) has gained significant attention both by industry and academia in the past years. The result is a proliferating development of ontologies, which significantly overlap [1,2] and, thus, recreate separated silos of associated knowledge. As Euzenat & Shvaiko [3] state: 'Thus, merely using ontologies, like using XML, does not reduce heterogeneity: it raises heterogeneity problems to a higher level'.

To address the reappearance of interoperability issues, there is a strong need to develop core ontologies to describe concepts and relationships reoccuring in every domain, e.g. the Building Topology Ontology (BOT) [1,4] and define alignments to it. The manual definition of alignments among domain ontologies is possible [2], however, almost as difficult as creating ontologies from scratch. Hence, there is an urgent need to [5]:

* Deploy existing ontology matching methods and tools in the AEC/FM domain  
* Create benchmarks from AEC/FM domain for ontology matching experts  

The tasks in this coding challenge involve:

* Deploy, fine tune and explore existing ontology matching methods and tools
* Revise and extend a set of existing ontologies towards developing a benchmark for the [Ontology Alignment Evaluation Initiative (OAEI)](http://oaei.ontologymatching.org/)


### Research Questions

* RQ1: Which levels of alignment occur when matching domain ontologies of the AEC/FM to BOT?
* RQ2: Which dedicated matching methods exist for the matching problems occuring in the AEC/FM domain?
* RQ3: Which domains and potential candidate ontologies need to be included to serve as a comprehensive benchmark for the OAEI?

* [BRICK2BOT](https://github.com/w3c-lbd-cg/bot/blob/AlignmentRevision/BRICKAlignment.ttl)  
* [DERIROOM2BOT](https://github.com/w3c-lbd-cg/bot/blob/AlignmentRevision/DERIROOMAlignment.ttl)  
* [DOGONT2BOT](https://github.com/w3c-lbd-cg/bot/blob/AlignmentRevision/DOGONTAlignment.ttl)

* [SAREF4BLDG2BOT](https://github.com/w3c-lbd-cg/bot/blob/AlignmentRevision/SAREF4BLDGAlignment.ttl)
